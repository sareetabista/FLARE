```python
scope.trigger.triggers = "tio4"
scope.adc.timeout = 10
```

```python
scope.arm()
ret = scope.capture()
print(ret)
```

```python
success = 0
fail = 0

for i in range(20):
    scope.arm()
    ret = scope.capture()
    if ret == False:
        success += 1
    else:
        fail += 1

print("trigger detected:", success)
print("timeouts:", fail)
```

```python
import chipwhisperer as cw
import serial
import time

UART_PORT = "/dev/ttyUSB1"
BAUD = 115200

scope.trigger.triggers = "tio4"
scope.adc.samples = 5000
scope.adc.timeout = 5

ser = serial.Serial(UART_PORT, BAUD, timeout=1)
time.sleep(1)

print("Starting synchronized capture test...")

detected = 0
timeouts = 0

for i in range(20):
    scope.arm()
    ret = scope.capture()

    line = ser.readline().decode(errors="ignore").strip()

    if ret is False:
        detected += 1
        print(f"[{i}] Trigger OK | UART: {line}")
    else:
        timeouts += 1
        print(f"[{i}] TIMEOUT | UART: {line}")

print("Done")
print("trigger detected:", detected)
print("timeouts:", timeouts)

ser.close()
scope.dis()
```
